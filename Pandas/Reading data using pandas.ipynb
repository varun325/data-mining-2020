# Filtering of the data

```python
df.loc[df['Type 1']=='Grass'] #single condition
```

```python
df.loc[(df['Type 1']=='Grass')&(df['Type 2']=='Poison')]#multiple conditions with and
```

```python
df.loc[(df['Type 1']=='Grass')|(df['Type 2']=='Poison')]#multiple conditions with or
```

```python
df.loc[((df['Type 1']=='Grass')|(df['Type 2']=='Poison'))&(df['HP']>80)]#multiple conditions
```

```python
new_df = df.loc[((df['Type 1']=='Grass')|(df['Type 2']=='Poison'))&(df['HP']>80)]#remove index
new_df = new_df.reset_index(drop=True)##drop=True drops the old index
#new_df.reset_index(drop=True,inplace=True)##does inplace , no need for the assignment
new_df
```

```python
df.loc[df['Name'].str.contains("Mega")] #fetch all the pokemons which are mega
```

```python
df.loc[~df['Name'].str.contains('Mega')]##fetch all the pokemons which are not Mega
```

```python
import re
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]
#re.I ignores case
#regex=True enables regex for the contains
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]+',flags=re.I,regex=True)]
```

# Aggregate statistics(Group by)

```python
df = pd.read_csv('pokemon_data.csv')
```

```python
df.groupby(['Type 1']).mean() # find the mean of all the values based upon Type 1
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)#sort in the descending 
#order of defence 
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack',ascending=False)#sort in the descending 
#order of attack
```

```python
df.groupby(['Type 1']).mean().sort_values('HP',ascending=False)#sort in the descending 
#order of attack
```

```python
df.groupby(['Type 1']).count()#count all the columns values horizontally for all the rows
```

```python
df['count']=1
df.groupby(['Type 1']).count()['count']#count the number of type 1
```

```python
df.groupby(['Type 1','Type 2']).count()['count']#count type 2 and type 1 combined
```

```python
chunks = pd.read_csv("pokemon_data.csv",chunksize=10)#load the data in the form of chunks
pieces = [x.groupby(['Type 1']).mean() for x in chunks]#save the chunks in pieces(list)
#get a list of pieces
new_df = pd.concat(pieces)#concatinate all the pieces
new_df.groupby(['Type 1']).mean()#get the mean values from the concatenated results
```
